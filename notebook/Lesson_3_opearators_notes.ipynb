# Lesson 2 – Variables & Data Types
## 1. What is a Variable?
A **variable** is like a container that stores a value in memory.
In Python, you don't need to declare the type — it's decided automatically.
```python
name = "Alice"     # Stores a string
age = 25           # Stores an integer
height = 5.6       # Stores a floating-point number
is_student = True  # Stores a boolean
```
---
## 2. Rules for Naming Variables
- ✅ Can contain letters, numbers, and underscores `_`
- ✅ Must start with a letter or `_`
- ❌ Cannot start with a number
- ❌ Cannot use reserved keywords (`if`, `for`, `class`, etc.)
---
## 3. Checking Data Type
```python
print(type(name))     # <class 'str'>
print(type(age))      # <class 'int'>
print(type(height))   # <class 'float'>
print(type(is_student)) # <class 'bool'>
```
---
## 4. Type Conversion (Casting)
Convert between data types:
```python
age_str = str(age)   # int → str
height_int = int(height)  # float → int
```

# Lesson 3 – Operators
## 1. Arithmetic Operators
| Operator | Meaning | Example |
|----------|---------|---------|
| `+` | Addition | `5 + 3` → 8 |
| `-` | Subtraction | `5 - 3` → 2 |
| `*` | Multiplication | `5 * 3` → 15 |
| `/` | Division | `5 / 2` → 2.5 |
| `//` | Floor Division | `5 // 2` → 2 |
| `%` | Modulus (remainder) | `5 % 2` → 1 |
| `**` | Exponent (power) | `2 ** 3` → 8 |
---
## 2. Comparison Operators
| Operator | Meaning |
|----------|---------|
| `==` | Equal to |
| `!=` | Not equal to |
| `>`  | Greater than |
| `<`  | Less than |
| `>=` | Greater than or equal to |
| `<=` | Less than or equal to |
```python
print(5 > 3)   # True
print(5 == 3)  # False
```
---
## 3. Logical Operators
| Operator | Meaning |
|----------|---------|
| `and` | True if both are True |
| `or`  | True if at least one is True |
| `not` | Opposite (True → False, False → True) |
```python
x = 5
print(x > 2 and x < 10)  # True
print(x > 2 or x > 10)   # True
print(not(x > 2))        # False
```
---
## 4. Assignment Operators
| Operator | Meaning | Example |
|----------|---------|---------|
| `=`  | Assign value | `x = 5` |
| `+=` | Add and assign | `x += 3` |
| `-=` | Subtract and assign | `x -= 2` |
| `*=` | Multiply and assign | `x *= 4` |
| `/=` | Divide and assign | `x /= 3` |
