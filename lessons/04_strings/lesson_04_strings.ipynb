# Lesson 4 – Strings
## 1. Introduction to Strings
Strings are sequences of characters enclosed in single quotes `'` or double quotes `"`.
```python
name = 'Python'
language = "Python"
```
---
## 2. String Indexing
- Strings are **indexed** starting from `0`.
- You can access characters by their position.
```python
word = "Python"
print(word[0])  # P
print(word[5])  # n
```
---
## 3. String Slicing
Extract part of a string:
```python
text = "Python Programming"
print(text[0:6])    # Python
print(text[7:])     # Programming
print(text[:6])     # Python
print(text[-6:])    # mming
```
---
## 4. Common String Methods
```python
msg = "  hello world  "
print(msg.upper())    # HELLO WORLD
print(msg.lower())    # hello world
print(msg.strip())    # hello world (removes spaces)
print(msg.replace("world", "Python"))  # hello Python
print(msg.title())    # Hello World
```
---
## 5. String Searching
```python
txt = "Python is fun"
print("fun" in txt)      # True
print(txt.find("is"))   # 7
```
---
## 6. f-Strings (Formatting)
```python
name = "Alice"
age = 25
print(f"My name is {name} and I am {age} years old.")
```
---
## 7. Escape Characters
```python
print("Hello\nWorld")   # New line
print("Hello\tWorld")   # Tab
print("She said \"Hi\"") # Quotes inside string
```
